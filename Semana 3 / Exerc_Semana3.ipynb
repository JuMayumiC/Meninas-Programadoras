# **Soletrando: atÃ© acertar!**

**Problema**

A professora do Ensino Fundamental gostou tanto do programa que vocÃª fez para o problemaÂ **Soletrando**Â que pediu para vocÃª fazer uma nova versÃ£o!

Ela quer que, em caso de erro em uma tentativa de soletrar a palavra, as crianÃ§as possam realizar novas tentativa elas mesmas! Para isso, vocÃªs combinaram com as crianÃ§as que elas podem tentar quantas vezes quiserem, e que:

- quando elas acertarem, o programa imprime **8-)** e termina

- quando elas errarem, o programa imprime **8-|** e elas podem tentar a mesma palavra outra vez

**Entrada**

- Primeira linha: uma palavra simples  (sem espaÃ§os)

- Linhas seguintes: uma letra por linha, terminadas com ponto.

- Se as letras corresponderem corretamente Ã  palavra da primeira linha, seu programa imprime **8-)**Â e termina

- Caso contrÃ¡rio,Â Â seu programa imprimeÂ **8-|** e lÃª um novo conjunto de linhas terminado por ponto.

**SaÃ­da**

- uma sequencia de zero ou mais linhas comÂ **8-|**

- ultima linha: **8-)**

Dicas

1) Resolva antes o **Soletrandro...**

2) Resolva antes o **Infinito enquanto dure**

3) Veja como esses dois ajudam na soluÃ§Ã£o desteÂ 

```python
word = input()
typed_word = ""

while True:
    letter = input()

    if letter == '.':
        print('8-|')
        typed_word = ""
    else:
        typed_word += letter

        if typed_word == word:
            print('8-)')
            break


```

**Qual problema vocÃª tem que resolver?**

Seu programa recebe os pontos de um set, um por linha: o valor 1 indica ponto do time 1, e o valor dois indica ponto do time 2. 

Seu programa deve aceitar pontos atÃ© que o set acabe, e imprimir o placar final do set.

**Dica**: projete sua soluÃ§Ã£o limitando o set a um numero de pontos bem menor que 25, por exemplo 5, mas mantendo a diferenÃ§a mÃ­nima de dois pontos para vitÃ³ria no set... depois de pronto, Ã© sÃ³ mudar a pontuaÃ§Ã£o para 25Â  >;-)

# **Problemas aceitos**

**Problemas aceitos**

A prova de nosso curso tem 4 problemas. Durante a prova, cada aluna pode submeter quantas vezes quiser uma soluÃ§Ã£o para cada problema desde que o tempo da prova nÃ£o tenha acabado.

 Seu programa deve identificar quais problemas foram aceitos antes do tempo terminar.

**Entrada**

Um sequencia de resultados de submissÃ£o que pode indicar resposta errada ou que o problema indicado foi aceito, como ilustrado nos exemplos. A entrada termina quando o final da prova Ã© indicado.Â 

**SaÃ­da**

Yes ou No para cada um dos quatro problemas, como ilustrado nos exemplos



```python
results = [0, 0, 0, 0]  # Inicializa uma lista para armazenar os resultados dos problemas
submission = input()  # LÃª a primeira submissÃ£o

while submission != "timeout":
    parts = submission.split()  # Separa a submissÃ£o em partes

    if len(parts) == 2:
        result, problem = parts  # ObtÃ©m o resultado e o nÃºmero do problema
        if result == "accepted":
            try:
                problem = int(problem) - 1  # Converte o nÃºmero do problema para um Ã­ndice vÃ¡lido na lista
                if 0 <= problem <= 3:
                    results[problem] = 1  # Marca o problema como aceito na lista
            except ValueError:
                pass  # Ignora a submissÃ£o invÃ¡lida se o nÃºmero do problema nÃ£o for um inteiro
    submission = input()  # LÃª a prÃ³xima submissÃ£o

# Imprime os resultados em uma Ãºnica linha
print(" ".join(["Yes" if result == 1 else "No" for result in results]))


```

O programa abaixo desenha um "histograma" na tela como base em valores lidos. Note que nosso histograma tem duas dimens√µes: 
> uma dimens√£o para cada linha do histograma

> para cada linha, a segunda dimens√£o representa qual o valor correspontente √†quela linha

Nosso programa:
* l√™ uma linha sequ√™ncia de valores da entrada, considera os valores como inteiros, e gera uma lista com esses valores
* para cada um dos valores da lista

```python
lista = [int(i) for i in input().split()]
for i in lista:
  linha = ''
  for j in range(i):
    linha += '*'
  print(linha)
```

#**Trabalhando com duas dimensões**

Neste exemplo, imprimimos um por linha cada item das duas dimensões

```python
dim = int(input())
for i in range(dim):
  print('*')
  for j in range(dim):
    print('-')
print('fim')
```

Neste exemplo, utilizamos o parâmetro **end** para explicitar para a função print() que ela não deve imprimir nada entre os itens das duas dimensões 

```python
dim = int(input())
for i in range(dim):
  print('*', end='')
  for j in range(dim):
    print('-', end='')
print('fim')
```

Neste exemplo, incluímos um caractere de nova linha **\n** antes da palavra **fim**

```python
dim = int(input())
for i in range(dim):
  print('*', end='')
  for j in range(dim):
    print('-', end='')
print('\nfim')
```

Neste exemplo, usamos o print() para inserir um final de linha ao final de cada linha da segunda dimensão

```python
dim = int(input())
for i in range(dim):
  for j in range(dim):
    print('*', end='')
  print()
print('fim')
```

E neste outro exemplo, o que está acontecendo? por quê?

```python
dim = int(input())
for i in range(dim):
  for j in range(dim):
    if i == j:
      print(' ', end='')
    else:
      print('*', end='')
  print()
print('fim')
```

E neste exemplo, o que está acontecendo? por quê?

```python
dim = int(input())
for i in range(dim):
  for j in range(dim):
    if i + j == dim:
      print(' ', end='')
    else:
      print('*', end='')
  print()
print('fim')
```

Olha como quem manda é a gente! 

```python
dim = int(input())
for i in range(dim):
  for j in range((i)):
    print('*', end='')
  print()
print('fim')
```

Olha como quem manda é a gente neste também! 

```python
dim = int(input())
for i in range(dim):
  for j in range((dim-i)):
    print('*', end='')
  print()
print('fim')
```

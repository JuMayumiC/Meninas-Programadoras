#**Verificar se um caractere pré-conhecido existe em uma string**

Depois de ler uma string, verificar se um caractere pré-conhecido existe nessa string.

Neste exemplo em que procuramos uma string específica, a letra 'a', em uma string lida e que fica apontada pela variável *nome*.

```python
nome = input()
if 'a' in nome:   # verifica se 'a' está contida em nome
  print('contém a')
else:
  print('não contém a')  
```

#**Verificar se uma substring lida está contida em uma string lida**

Naturalmente, a gente pode ler também a string que queremos verficar se existe na outra string.

```python
original = input()
parte = input()
if parte in original: # verifica se parte está contida em orignal
  print(original, 'contém', parte)
else:
  print(original, 'não contém', parte)
```

#**Desconsiderando a capitalização com lower()**

Note, no exemplo acima, que o comando condicional avalia a existência da substring *parte* na string completa *original* considerando a capitalização original (i.e., diferenciando letras minúsculas e maiúsculas), o que pode não ser o que queremos em algumas situações. 

Uma alternativa é compararmos versões das strings *original* e *parte* cujas letras são todas consideradas letras minúsculas usando o método lower().






```python
original = input()
parte = input()
if parte.lower() in original.lower(): # compara os valores convertidos usando lower() sem sobrescrever as versões originais
  print(original, 'contém', parte)
else:
  print(original, 'não contém', parte)
```

#**Desconsiderando a capitalização com upper()**

Obtemos resultado similar se utilizarmos o método upper(), *pois* comparamos versões temporárias que utilizam a mesma capitalização nos caracteres individuais

```python
original = input()
parte = input()
if parte.upper() in original.upper(): # compara os valores convertidos usando upper() sem sobrescrever as versões originais
  print(original, 'contém', parte)
else:
  print(original, 'não contém', parte)
```

#**Alguns outros métodos que podemos utilizar com strings**

Alguns outros métodos disponíveis são:
* count() retorna quantas vezes a substring passada como argumento existe na string original
* isalnum() retorna True se todos os caracteres da string são alfanuméricos
* isalpha() retorna True se todos os caracteres da string são alfabéticos
* isdigit() retorna True se todos os caracteres da string são dígitos

```python
print("banana".count('na'))
print("oftalmotorrinolaringologista".count('o'))

print("men1nas".isalnum())
print("meninas".isalnum())

print("men1nas".isalpha())
print("meninas".isalpha())

print("men1nas".isdigit())
print("5647282910".isdigit())
```

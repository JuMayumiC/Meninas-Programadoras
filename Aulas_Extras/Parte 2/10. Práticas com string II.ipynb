#**Faça você mesma: como você faria um contador?**

Podemos utilizar o comando for-in para analisar uma letra da string de cada vez.

Para contar, precisamos utilizar uma variável que vai ser incrementada em uma unidade toda vez a comparação for True.
* note que no programa utilizamos um atalho para fazer a atualização do contador
* esse atalho funciona também para os demais operadores, como nos exemplos 
```
    contador += 1       # atalho para contador = contador + 1
    contador -= 1       # atalho para contador = contador - 1
    contador *= 2       # atalho para contador = contador * 2
    contador /= 3       # atalho para contador = contador / 3
```



````python
original = input()      # string onde queremos procurar um caractere
caractere = input()     # caractere a ser procurado
contador = 0            # começamos a contar a partir do zero

for letra in original:
  if letra == caractere: 
    contador += 1       # atalho para contador = contador + 1
print(contador)
````

#**Faça você mesma: como você faria um contador de vogais?**

* Podemos utilizar o comando for-in para analisar uma letra da string de cada vez.
* Em vez de comparar com uma letra, podemos comparar com um conjunto de letras.



````python
original = input()  
vogais = input()
contador = 0      # começamos a contar a partir do zero

for letra in original:
  if letra in vogais: 
    contador += 1       # atalho para contador = contador + 1
print(contador)
````

#**Faça você mesma: como você faria um mascarador de texto?**

Às vezes queremos mascarar um texto de modo a dificultar a identificação do conteúdo por pessoas não conhecem o texto original.

* Podemos utilizar o comando for-in para analisar uma letra da string de cada vez.
* Em vez de comparar com uma letra, podemos comparar com um conjunto de letras.

* Precisamos montar a substring resultante
>* começamos com a string vazia
>* acrescentamos um caractere de cada vez
>* antes de acrescentar, verificamos se faz parte do conjunto que temos que mascarar 

````python
original = input()
trocar = input()
resultado = ''
for letra in original:
  if letra in trocar:
    resultado += '*'
  else:
    resultado += letra
print(resultado)
````

#**Faça você mesma: como você faria um separador de cpf?**

Todo CPF tem 11 digitos, e.g. 12345678900

Normalmente, os dígitos de um CPF são separados por pontos e um hífen para facilitar a leitura por seres humanos, e.g. 123.456.789-00

Dado o conjunto de dígitos de um CPF, como gerar uma versão com os separadores?

Note que:
* precisamos colocar um ponto **'.'** depois do terceiro e do sexto dígito
* precisamos colocar um hífen **'-'** depois do nono dígito

Além disso, note que:
* para essa tarefa, tratamos o CPF como um conjunto de caracteres. 
* nosso programa *confia* que o valor fornecido pelo usuário é um conjunto de dígitos correspondente a um CPF válido.

````python
cpf = input()   # não convertemos para inteiro pois tratamos o valor com cadeia de caracteres
resultado = ''  # vai apontar para o CPF com os separadores
contador = 0    # vai controlar quando devemos inserir os separadores
for digito in cpf:
  resultado += digito                 # somamos strings aqui!
  contador += 1                       # avançamos com o contador
  if contador == 3 or contador == 6:
    resultado += '.'
  elif contador == 9:
    resultado += '-'
print(resultado)
````

#**Faça você mesma: como você faria um separador de cpf mascarado?**

E se a gente quiser mascarar o CPF também? 

Uma alternativa é a gente considerar a entrada como sendo os 11 caracteres originais e
* primeiro trocamos cada *posição par* da entrada por um asterisco
* depois colocamos os separadores no cpf já mascarado

````python
cpf = input()   # não convertemos para inteiro pois tratamos o valor com cadeia de caracteres

# colocamos as máscaras

resultado = ''  # vai apontar para o CPF com os separadores
contador = 0    # vai controlar quando devemos inserir os separadores
for digito in cpf:
  contador += 1           # avançamos com o contador
  if contador % 2 == 0:   # se for casa par
    resultado += '*'
  else:
    resultado += digito
print(resultado)

# reinicializamos cpf, resultado e contador
cpf = resultado
resultado = ''
contador = 0

# colocamos os separadores

for digito in cpf:
  resultado += digito                 # somamos strings aqui!
  contador += 1                       # avançamos com o contador
  if contador == 3 or contador == 6:
    resultado += '.'
  elif contador == 9:
    resultado += '-'
print(resultado)
````

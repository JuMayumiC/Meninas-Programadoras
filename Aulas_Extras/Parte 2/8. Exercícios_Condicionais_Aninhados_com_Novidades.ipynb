#**Eventos condicionais encadeados**
Em qualquer bloco de instruções, podemos incluir quais comandos válidos. Por isso, é possível incluir comandos condicionais dentro de outros comandos condicionais, e não há limite para o aninhamento.

#**Exercício**

Considere um programa que a ser implementado em um totem disponibilizado em um salão de eventos. O programa do tótem deve informar que o evento é apenas para pessoas do gênero feminino e crianças que vierem com elas. Se a pessoa for mãe, o totem deve perguntar o nome dos filhos. Para toda mulher, o totem deve dar as boas vindas e, se for mãe, informar para trazer os filhos citando o nome deles e, se não for, dizer que a pessoa pode trazer alguma criança.

```
ALGORITMO
obter gênero
se gênero feminino
  obter nome
  prepara resposta de boas vindas
  obter se é mãe ou não
    se for mãe
       obter nome dos filhos
       atualiza resposta para trazer os filhos, citando os nomes
    senao
       atualiza resposta para trazer alguma criança
senao
  prepara resposta para evento para público feminino
apresenta resposta
```

**Elementos conhecidos**
* leitura de inteiros
* strings constantes demarcadas com aspas simples
* operações de multiplicação e subtração de inteiros
* operação de soma de strings
* condicional if-elif-else
* impressão de string

**Novidade**
* leitura dentro de comando condicional: os dados que são lidos dependem de respostas anteriores
* comandos if-else aninhados

**Faça você mesma**
* mude o exercício para realizar outras perguntas e outros testes
* apresente o resultado de maneiras diferentes

````python
genero = input('gênero: F ou M? ')
if genero == 'F':
  nome = input('nome: ')
  resposta = 'Bem-vinda ' + nome + '!\n'
  mae = input('mãe: S ou N?')
  if mae == 'S':
    filhos = input('nome dos filhos: ')
    resposta += 'Para o evento, traga seus filhos: ' + filhos
  else:
    resposta += 'para o evento, traga alguma criança'
else:
  resposta = 'Evento apena para pessoas do gênero Feminino'
print(resposta)
````

#**Exercício**

Considere o programa de uma malha de esteiras de análise de alimentos. 
* Na primeira esteira, podem passar alimentos de dois tipos identificados por um sensor (e.g. alhos e bugalhos). Ao final da esteira há uma bifurcação, e o programa deve encaminar para o ramo correto da bifurcação.
* No segundo nível de esteiras, o programa recebe dado de outro sensor que identifica dois outros aspectos (e.g. alhos grandes e alhos pequenos, e bugalhos moles e bugalhos duros). Novamente, ao final de cada esteira há uma bifurcação que deve ser acionada corretamente pelo programa. 


```
ALGORITMO
ler sensorAB
se tipo == alho
  liberar esquerda
  ler sensorGP
  se tipo = grande
    liberar esquerda
  senao # pequeno
    liberar direita
senao  # bugalho
  liberar direita
  ler sensorMD
  se tipo = mole
    liberar esquerda
  senao  # duro
    liberar direita
```


**Elementos conhecidos**
* leitura de inteiros
* strings constantes demarcadas com aspas simples
* operações de comparação de strings
* condicional if-elif-else
* impressão de string
* leitura dentro de comando condicional: os dados que são lidos dependem de respostas anteriores
* comandos if-else aninhados

**Novidade**
* bugalhos de vários tipos :-)

**Faça você mesma**

* apresente o resultado de maneiras diferentes

````python
sensor = input()
if sensor == 'alho':
  print('vá para a esquerda')
  sensor = input()
  if sensor == 'grande':
    print('vá para a esquerda')
  else: # pequeno
    print('vá para a direita')
else: # bugalho
  print('vá para a direita')
  sensor = input()
  if sensor == 'mole':
    print('vá para a esquerda')
  else: #duro
    print('vá para a direita')
````

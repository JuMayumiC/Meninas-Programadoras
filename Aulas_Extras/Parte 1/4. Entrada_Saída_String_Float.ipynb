#**Argumentos da função print()**

A função print() recebe zero ou mais argumentos que indicam o que desejamos imprimir em nosso programa.
* uma chamada sem argumentos imprime uma linha em branco (ou fim de linha)
* uma camada com um argumento, imprime o valor correspondente, seguido de um fim de linha
* uma camada com mais que um argumento, imprime os valores correspondentes separados por um espaço em branco, seguido de um fim de linha

````python
texto1 = input()
texto2 = input()
print('-----')
print()
print(texto1)
print(texto2)
print(texto1, texto2)
````

#**Imprimindo valores decimais: um modo simples**

Na vida real, muitas vezes não queremos utilizar todas as casas decimais resultantes de uma operação. 

Um exemplo é quando o resultado deve ser um valor inteiro: 
* não é possível pedir a resolução de _pelo menos 7.5 exercícios_ de uma lista
* não é possível aceitar 33.33333% das alunas inscritas no curso.
* nesses casos devemos arredondar o valor decimal obtido para conter zero casas decimais

Podemos passar para a funçao print() argumentos especificam com quantas casas decimais um valor dever ser apresentado. Por exemplo:

```
print('%.2f' % x)
```

Nesse exemplo, o argumento passado para a função print() segue a sintaxe:
* sequência de caracteres (entre aspas) que contém um símbolo **%**
* outro símbolo **%** 
* e o nome de uma variável

A sequência de caracteres (entre aspas), por sua vez, segue a sintaxe:
* o símbolo **%** 
* seguido do símbolo **.** 
* seguido de um número inteiro 
* seguido da letra **f** 

Os argumentos **'%.2f' % x** formam uma expressão que informa ao interpretador Python que o valor apontado pela variável **x** deve ser apresentado com duas casas decimais:
* o número *2* indica que o valor apontado por *x* deve ser impresso com **duas casas** 
* e o símbolo **f** indica que o valor deve ser considerado um float(), ou seja, com casas decimais.
                                                            

````python
decimal = float(input())
print('%.3f' % decimal)
print('%.2f' % decimal)
print('%.1f' % decimal)
print('%.0f' % decimal)
````

#**Apresentando uma mensagem ao pedir um valor de entrada**
É possível apresentar uma mensagem de texto ao usuário quando solicitamos um valor de entrada. Isso é muito útil em programas interativos.


````python
nome = input('Seu nome:')
sobrenome = input('Sobrenome:')
print('Bem-vinda', nome, sobrenome, '!')
````

#**Uma utilidade para concatenação de sequências de caracteres**
Note no exemplo acima que o resultado da função print() sempre apresenta um espaço do ponto de exclamação. Por que isso acontece?
* porque a função print() coloca um espaço entre entre os argumentos que estão separados por vírgula

Quando não queremos que esse espaço seja incluído, uma alternativa é criar a string que queremos imprimir separadamente, e então passar essa string como argumento para a função print(). 
* nesse caso, é de nossa inteira responsabilidade incluir todos os espaços desejados entre o conteúdo referenciado pelas variáveis que utilizamos na composição da string. 

````python
nome = input('Seu nome:')
sobrenome = input('Sobrenome:')
saida = 'Bem-vinda ' + nome + ' ' + sobrenome +'!'
print(saida)
````

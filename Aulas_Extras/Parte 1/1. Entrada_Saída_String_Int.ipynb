#**Enviando informações para o mundo**
Utilizamos a função print() para passar conteúdo do nosso programa para o mundo.
Dentro dos parênteses colocarmos diretamente os valores que queremos mostrar.
Quando temos que passar mais de um valor, indicamos o conteúdo entre vírgulas. A função print() inclui automaticamente um espaço entre os conteúdos. 

```python
print("Meninas Programadoras")
print('Curso de Extensão')
print("Meninas Programadoras", 'Curso de Extensão')
```

#**Obtendo dados do mundo**
Para obter informações do mundo para o nosso programa, utilizamos a função input(). Essa função lê uma sequência de caracteres da entrada padrão e armazena em algum lugar na memória do computador. Para podermos utilizar esse valor em nosso programa, temos que saber onde ele foi armazenado. Para isso, a gente utiliza uma variável que, ao receber o resultado da função input(), sabe onde o conteúdo foi armazenado.

O nome de uma variável deve seguir algumas regras. Em nossos programas, vamos seguir este subconjunto de regras: 
* sempre começar com uma letra: vamos usar sempre letras minúsculas 
* usar apenas letras e números
* não podemos incluir espaço: utilizamos _underscore_ como separador 
vamos começar com letra minúsculas

Além disso, é importante sempre usar nomes significativos para as variáveis. Isso ajuda nós mesmas a ler o que o programa deve fazer.

```python
sequencia_de_caracteres = input()
nome = input()
sobrenome = input()
cidade = input()
estado = input()
print(sequencia_de_caracteres)
print(nome, sobrenome)
print(cidade, estado)
```

#**Obtendo valores inteiros do mundo**
A função **input()** lê uma sequência de caracteres da entrada padrão e armazena em algum lugar na memória do computador. Quando queremos que aqueles valores sejam reconhecidos em nossos programas como valores inteiros ou decimais, precisamos explicitar isso no programa. Se não fizermos isso, as operações sobre valores inteiros e decimais não terão o resultado desejado. 

Veja o erro que acontece no exemplo a seguir:

```python
valor = input()
valor_seguinte = valor + 1
print(valor_seguinte) 
```

#**Obtendo valores inteiros e decimais do mundo**

Portando, quando precisamos que valores obtidos pela função **input()** sejam reconhecidos em nossos programas como valores inteiros ou decimais, precisamos de uma forma de explicitar essa necessidade em nosso programa. 

> Utilizamos a função **int()** para transformar um valor do tipo sequência de caracteres em um valor inteiro.

> Utilizamos a função **float()** para transformar um valor do tipo sequência de caracteres em um valor com casas decimais. 

* Note que não precisamos incluir o ponto decimal ou casas decimais quando oferecemos o valor ao programa.


```python
valor1 = input() # coloca primeiro conteúdo obtido na memória
valor2 = input() # coloca segundo conteúdo obtido na memória
valor1 = int(valor1) # o primeiro conteúdo passa a ser considerado um inteiro
valor2 = float(valor2) # o segundo conteúdo passa a ser considerado um inteiro
print(valor1 + 1)  # podemos fazer uma operação aritmética entre valores inteiros
print(valor2 + 1.5)  # podemos fazer uma operação aritmética entre valores inteiros
```

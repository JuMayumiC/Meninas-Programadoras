#**Operadores de comparação**
* O valor 10 é maior que o valor 8?
* O valor -1 é menor que o valor -3?
* O valor 0 é igual ao valor -100?


Quando comparamos dois valores, qual o resultado?
O resultado da comparação é **verdadeiro** ou **falso**, certo?

Por exemplo, quando comparamos 
* 1 > 2
* 1 < 2
* 1 == 2
* 1 >= 2
* 1 <= 2

Os resultados são, respectivamente
* falso
* verdadeiro
* falso
* falso
* verdadeiro

Em Python, os valores verdadeiro e falso são, respectivamente, **True** e **False**.


```python
valor1 = int(input())
valor2 = int(input())
resposta = valor1 > valor2
print(resposta)
resposta = valor1 >= valor2
print(resposta)
resposta = valor1 < valor2
print(resposta)
resposta = valor1 <= valor2
print(resposta)
resposta = valor1 == valor2
print(resposta)
```

#**Operadores lógicos**

Como na vida real, muitas vezes uma condição depende de outras condições. Por exemplo:
* (a > b) **e** (c < d)
* (a > b) **ou** (c < d)
* **não** (a > b) 

Os operadores em Python são todos em inglês: **and**,  **or** e **not** 



```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())
resposta = (a > b) and (c < d) 
print(resposta)
resposta = (a > b) or (c < d) 
print(resposta)
resposta = not (a > b) 
print(resposta)
```

#**Operador aritmético de adição**
O tempo todo precisamos saber que tipo de valores as nossas variáveis apontam na memória do computador. Com isso, dominamos o resultado de operações que utilizamos em nosso programa.

Um exemplo é o operador de soma. O resultado da adição de dois valores depende do tipo dos dois valores, certo?
* a soma de dois valores inteiros resulta em um valor inteiro
* a soma de dois valores decimais resulta em um valor decimal
* a soma de um valor inteiro e um valor decimal resulta em um valor decimal
* e a soma de dois valores do tipo sequência de caracteres resulta em uma nova sequência de caracteres



```python
inteiro = int(input())
decimal = float(input())
sequencia_de_caracteres = input()
soma_inteiro = inteiro + 1
soma_decimal = decimal + 1
soma_inteiro_decimal = inteiro + decimal
soma_sequencia_de_caracteres = sequencia_de_caracteres + '!!!'
print(soma_inteiro)
print(soma_decimal)
print(soma_inteiro_decimal)
print(soma_sequencia_de_caracteres)
```

#**Operador aritmético de multiplicação**

Outro exemplo é o operador de multiplicação. O resultado da multiplicação de dois valoresd depende do tipo dos dois valores, certo?
* a multiplicação de dois valores inteiros resulta em um valor inteiro
* a multiplicação de dois valores decimais resulta em um valor decimal
* a multiplicação de um valor inteiro e um valor decimal resulta em um valor decimal

Detalhe: 
* a multiplicação de um valor do tipo sequência de caracteres por um valor inteiro em uma nova sequência de caracteres

```python
inteiro = int(input())
decimal = float(input())
sequencia_de_caracteres = input()
multiplicacao_inteiro = inteiro * 10
multiplicacao_decimal = decimal * 100
multiplicacao_inteiro_decimal = inteiro * decimal
multiplicacao_sequencia_de_caracteres = sequencia_de_caracteres * 3
print(multiplicacao_inteiro)
print(multiplicacao_decimal)
print(multiplicacao_inteiro_decimal)
print(multiplicacao_sequencia_de_caracteres)
```

#**Operadores aritméticos sobre valores inteiros**
Em nossos programas, os operadores aritméticos funcionam como aprendemos desde pequenas. Em alguns casos o símbolo utilizado muda, principalmente para podermos escrever as operações em uma linha. Por exemplo, 
* a gente escreve *maior ou igual* com o símbolo do *maior* seguido do símbolo do *igual* ou seja **>=** 

Além disso, é útil a gente lembrar uma operação que aprendemos a fazer ainda crianças: a divisão entre valores inteiros.
* o operador **//**  indica a divisão de um **dividendo inteiro** por um **divisor inteiro** e o resultado é o **quociente inteiro** correspondente
* o operador **%** indica que queremos obter o resto da divisão de um **dividendo inteiro** por um **divisor inteiro**: o resultado é o **resto** da divisão


```python
inteiro1 = int(input())
inteiro2 = int(input())
resposta = inteiro1 + inteiro2 # soma
print(resposta)
resposta = inteiro1 - inteiro2 # subtração
print(resposta)
resposta = inteiro1 * inteiro2 # multiplicação
print(resposta)
resposta = inteiro1 ** inteiro2 # exponenciação
print(resposta)
resposta = inteiro1 // inteiro2 # divisor resultante da divisão inteira
print(resposta)
resposta = inteiro1 % inteiro2 # resto resultante da divisão inteira
print(resposta)
```

#**Operadores aritméticos envolvendo valores decimais**
Em uma operação que envolve dois operandos, quando um deles é decimal e resultado é decimal. 
Além disso, a quando utilizamos o operador de divisão / entre dois valores, estamos especificando que queremos o resultado da divisão decimal correspondente.
* Lembre que não precisamos incluir o ponto decimal ou casas decimais quando oferecemos o valor ao programa.

```python
inteiro1 = int(input())
inteiro2 = int(input())
decimal1 = float(input())
decimal2 = float(input())
resposta = inteiro1 / inteiro2
print(resposta)
resposta = decimal1 / decimal2
print(resposta)
```

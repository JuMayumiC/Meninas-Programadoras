#**Execução condicional** 
Em muitos de nossos programas, desejamos executar um bloco de instruções apenas se uma determinada condição for verdadeira **True**.

##**Execução condicional se verdadeiro (comando if)** 

Seja o exemplo de identificarmos se um determinado valor inteiro obtido da entrada é positivo. Execute o programa abaixo oferecendo:
* valores positivos
* valores negativos
* o valor nulo

````python
valor = int(input())
if valor > 0:
  print(valor, 'é positivo') # bloco interno ao if, indicado pelo deslocamento
print('fim.')                # comando seguinte ao bloco
````

Você notou que 
* o programa executa o _bloco interno ao_ **if** apenas quando o valor que você ofereceu foi positivo?
* em qualquer caso, o comando da **print('fim.')** é executado?

##O que acontece?
Quando utilizamos um comando condicional em nossos programas, especificamos que um conjunto de comandos só deve ser executado quando a condição que incluímos no comando for verdadeira.

## Sintaxe
O comando condicional **if** segue a seguinte sintaxe:
```
if condição:
   comando1
   comando2
   ...
   comandoN
```

Nessa sintaxe:
* a palavra reservada **if** indica o início do comando
* o conteúdo entre a palavra **if** e o símbolo de dois pontos **:** especifica uma condição lógica, isto é, sua avaliação deve resultar em uma valor **True** ou **False**
* um cojunto de um um mais comandos é apresentado de modo deslocado em relação à palavra **if**
* todos os comandos devem ter o mesmo deslocamento

 

````python
nome = input()                # lê nome
idade = int(input())          # lê idade
resposta = 'não pode votar'   # prepara resposta para 'não pode votar'
if idade >= 16:               # testa condição para idade de votante
  resposta = 'pode votar'     # se condição for verdadeira, corrige a resposta
print(nome, resposta)         # executado depois do bloco if
````

##**Execução condicional verdadeiro-falso (comando if-else)**
O exemplo acima ilustra uma situação em que, de fato, precisamos executar um bloco de instruções se a condição avaliada pelo **if**  resultar na condição verdadeira, e outro bloco de instruções se a condição for avaliada como falsa. 
* nesses casos, utilizamos a versão condicional **if-else** 

Neste caso, utilizamos a palavra reservada **else** 
* no mesmo alinhamento da palavra **if** correspondente 
* imediatamente depois do else utilizamos um dois pontos **:** pois a condição de execução do bloco correspondente está implícita
* o bloco correspondente ao **else** será executado sempre que a condição avaliada na parte **if** correspondente for avaliada como **False**

````python
nome = input()                # lê nome
idade = int(input())          # lê idade
if idade >= 16:               # testa condição para idade de votante
  resposta = 'pode votar'     # se condição for verdadeira, prepara resposta positiva
else:                         # identifica bloco executado se condição for negativa
  resposta = 'não pode votar' # prepara resposta negativa
print(nome, resposta)         # executado depois do bloco if-else
````

##**Par ou ímpar**
Seja o problema: _dado um valor inteiro, ele é par ou ímpar?_

Sabemos que:
* todo número par é divisível por 2

Portanto
* se um número é par, ele é divisível por 2, 
* caso contrário, ele é ímpar

Esse é um caso típico de utilizamos o comando **if-else**, certo?
* e como sabemos se um dado valor inteiro é divisível por 2?
* um valor _ser divisível por 2_ significa que o resto da divisão do valor por 2 é zero 

Agora é útil lembrarmos dos operadores associados à divisão inteira entre dois valores inteiros:
* o operador **//**  indica a divisão de um **dividendo inteiro** por um **divisor inteiro** e o resultado é o **quociente inteiro** correspondente
* o operador **%** indica que queremos obter o resto da divisão de um **dividendo inteiro** por um **divisor inteiro**: o resultado é o **resto** da divisão

````python
valor = int(input())
if valor % 2 == 0: 
  resposta = 'par'
else:
  resposta = 'ímpar'
print(valor, 'é', resposta)
````

##**Execução condicionada a várias alternativas (if-elif-else)**
Há casos em que precisamos executar blocos de comandos associados a mais que duas alternativas, e nesse caso podemos utilizar uma versão mais poderosa do comando condicional.   

````python
nome = input()
idade = int(input())
if idade < 16:
  resposta = 'é muito cedo para você votar'
elif idade < 18:
  resposta = 'seu voto facultativo'
elif idade < 70:
  resposta = 'seu voto obrigatório'
else:
  resposta = 'seu voto facultativo'
print(nome, 'sua idade indica que', resposta)
````

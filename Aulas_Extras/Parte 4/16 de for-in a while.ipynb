#**A gente sabe fazer o que o _if-in_ faz em uma lista: solução ingênua**

> varremos nós mesmas buscar em uma lista, comparando cada elemento com a chave

> neste exemplo, a gente **sempre** passeia na lista até o final, mesmo que ache a chave antes 8-(


```python
chave = int(input())
lista = [19, 9, -55, 99, 8, 7, 5, -9]

for i in lista:
  if i == chave:
    print('achei')
  else:
    print('nao é este aqui', i)
```

#**Agora a gente passeia na lista e para assim que possível com _break_**

> varremos nós mesmas buscar em uma lista, comparando cada elemento com a chave

> e a gente para da primeira vez que encontrar!


```python
chave = int(input())
lista = [19, 9, -55, 99, 8, 7, 5, -9]

for i in lista:
  if i == chave:
    print('achei')
    break
  else:
    print('nao é este aqui', i)
```

#**É um trabalho imenso para busca de uma chave em uma lista grande** 

* Primeiro a gente gera uma lista "grande"
* Depois fazemos nós mesmas a busca para entender bem o trabalhão que Python faz para a gente


```python
import random
OK = 'achei'
ERRO = 'nao tem'

# primeiro geramos uma lista para termos com o que trabalhar
lista = []
for i in range(9999):
  numero = int(random.uniform(1,99999))
  lista.append(numero)
print(lista)

chave = int(input())
resposta = ERRO
trabalhao = ''
for i in lista:
  if chave == i:
    resposta = OK
  else:
    trabalhao += '.'
print(resposta)
print(trabalhao)
```

#**A gente sabe fazer um contador genérico**

```python
# Ler primeiro e últimos elementos
# Imprimir todos os inteiros entre o primeiro e o último, inclusive o primeiro e o último
# e.g.
#  de 7    a   10
#     7  8  9  10
# neste, imprimimos todos os valores na mesma linha

inicio = int(input())
final = int(input())
contador = 0
while inicio <= final:
  print(inicio, end='-')
  contador += 1
  inicio += 1
```

#**A gente sabe fazer o que a função count() da classe lista faz**

```python
import random
OK = 'achei'
ERRO = 'nao tem'

# gerar uma lista para ter com o que trabalhar
lista = []
for i in range(2000):
  numero = int(random.uniform(1,999))
  lista.append(numero)
print(lista)

# obter  chave a ser buscada 
chave = int(input())

resposta = ERRO   # inicializar resposta para o caso de não encontrar
contador = 0      # inicializar contador 
for i in lista:   # mandar Python passear na lista
  if chave == i:  # compararar item com chave
    resposta = OK # atualizar resposta toda vez que encontrar
    contador += 1 # atualizar contador

print(resposta)
print('ocorrências:', contador)
```

#**A gente sabe fazer um jogo de chutar um número**

> um chute por execução

```python
import random

# gerar uma lista para ter com o que trabalhar
lista = []
for i in range(20):
  numero = int(random.uniform(1,20))
  lista.append(numero)

chute = int(input())          # obter chute do usuário
contador = lista.count(chute) # mandar Python contar quantos tem

if contador > 0:              # verificar se tem algum ou não
  print('achou')
else:
  print('dançou')

print(lista)
```

#**Podemos usar _while_ para deixar o usuário chutar até acertar**

> usar **while** para deixar o usuário chutar até que o valor seja encontrado

```python
import random

# gerar uma lista para ter com o que trabalhar
lista = []
for i in range(20):
  numero = int(random.uniform(1,20))
  lista.append(numero)

chute = int(input())          # obter chute do usuário
contador = lista.count(chute) # mandar Python contar quantos tem

while contador == 0:          # repetir até encontrar pelo menos 1
  print('tente de novo 8^)')
  chute = int(input())
  contador = lista.count(chute)
else:
  print('achou')

print(lista)
```

#**Podemos usar _while_ para testar até quando devemos repetir uma tarefa  até concluí-la**


```python
conta = int(input())
montinho = 0
while montinho < conta:
  valor = int(input())
  montinho += valor
print(montinho)
```

#**Podemos usar _while_ _TRUE_ e usar o break para encerrar a repetição quando a tarefa é concluída**


```python
import random

# primeiro geramos uma lista para termos com o que trabalhar
lista = []
for i in range(20):
  numero = int(random.uniform(1,20))
  lista.append(numero)

while True:
  chute = int(input())
  contador = lista.count(chute)
  if contador != 0:
    print('achou')
    break
  else:
    print('tente de novo 8^)')

print(lista)
```

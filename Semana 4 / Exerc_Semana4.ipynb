```python
n = int(input())
e = []

for _ in range(n):
    email = input()
    e += [email]  

count_invalid = 0

for email in e:
    if '@' not in email:
        count_invalid += 1

print(count_invalid)
```

# **Soma, subtraÃ§Ã£o, multiplicaÃ§Ã£o, divisÃ£o, mÃ³dulo, divisÃ£o inteira e potÃªncia**

**Vamos resolver**

Dona Maria quer que vocÃª responda rÃ¡pido: dados dois valores inteiros, qual o resultado das operaÃ§Ãµes matemÃ¡ticas soma, subtraÃ§Ã£o, multiplicaÃ§Ã£o, divisÃ£o, mÃ³dulo, divisÃ£o inteira e potÃªncia entre os dois? Como vocÃª nÃ£o tem calculadora, vocÃª decide escrever um programa que calcula as respostas rapidamente. Considere, quando necessÃ¡rio, valores com  duas casas decimais.

**Entrada**

Duas linhas: cada uma contÃ©m um valor inteiro

**SaÃ­da**

O resultado de sete operaÃ§Ãµes matemÃ¡ticas: soma, subtraÃ§Ã£o, multiplicaÃ§Ã£o, divisÃ£o, mÃ³dulo, divisÃ£o inteira e potÃªncia, nessa ordem, entre os dois valores fornecidos. 

- Um valor em cada linha.Â 

- Quando necessÃ¡rio, considere valores com  duas casas decimais.

```python

valor1 = int(input())
valor2 = int(input())


soma = valor1 + valor2
subtracao = valor1 - valor2
multiplicacao = valor1 * valor2
divisao = valor1 / valor2
modulo = valor1 % valor2
divisao_inteira = valor1 // valor2
potencia = valor1 ** valor2


print("{:.2f}".format(soma) if soma % 1 != 0 else str(int(soma)))
print("{:.2f}".format(subtracao) if subtracao % 1 != 0 else str(int(subtracao)))
print("{:.2f}".format(multiplicacao) if multiplicacao % 1 != 0 else str(int(multiplicacao)))
print("{:.2f}".format(divisao))
print("{:.2f}".format(modulo) if modulo % 1 != 0 else str(int(modulo)))
print("{:.2f}".format(divisao_inteira) if divisao_inteira % 1 != 0 else str(int(divisao_inteira)))
print("{:.2f}".format(potencia) if potencia % 1 != 0 else str(int(potencia)))
```

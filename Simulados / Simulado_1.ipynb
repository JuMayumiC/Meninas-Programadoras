# **SubmissÃµes e Cia**

**Problema**

Vamos ajudar a equipe do projeto Meninas Programadoras a saberem como as alunas resolveram uma lista de exercÃ­cio? VocÃª tem que fazer um programa para verificar, para uma aluna, quantas submissÃµes foram realizadas por ela, e quantas foram aceitas, no prazo de vigÃªncia da lista. 

O programa primeiro recebe o nÃºmero de problemas da lista. A seguir, o programa recebe um nÃºmero indefinido de linhas que podem conter as palavras 'accepted', 'error' ou 'timeout'. Quando ocorre 'timeout' nÃ£o sÃ£o aceitas novas submissÃµes.

**Entrada**

* primeira linha: nÃºmero de problemas
* a seguir: um nÃºmero indefinido de linhas que podem conter as palavras 'accepted', 'error' ou 'timeout'

**SaÃ­da**

Como nos exemplos


```python
ac = 0
subc = 0
np = int(input())
sub = input()

while True:
    subc += 1
    if sub == 'accepted':
        ac += 1

    if ac == np or sub == "timeout":
        break

    sub = input()

print('submissoes:', subc)
print('aceitos:', ac)
```

# **C*F mascarado**

Ã‰Â sempreÂ bomÂ mascararÂ informaÃ§ÃµesÂ paraÂ protegerÂ privacidadeÂ dosÂ donosÂ dasÂ informaÃ§Ãµes.Â 

SeuÂ programaÂ deveÂ lerÂ oÂ nÃºmeroÂ deÂ CPFÂ (apenasÂ nÃºmeros)Â eÂ gerarÂ umaÂ mascaraÂ trocandoÂ todosÂ osÂ valoresÂ paresÂ peloÂ asterisco.



**Entrada**

Os dÃ­gitos que compÃµem um CPF

**SaÃ­da**

O CPF mascarado como especificado acimaÂ Â 

```python
npar = [0, 2, 4, 5, 8]
cpf = input()

for i in npar:
    cpf = cpf.replace(str(i), '*')

print(cpf)
```

```python
cpf = input()
cpf_masked = ''

for digit in cpf:
    if int(digit) % 2 == 0:
        cpf_masked += '*'
    else:
        cpf_masked += digit

print(cpf_masked)
```
